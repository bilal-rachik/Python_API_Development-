## Comprendre L'API

API signifie interface de programmation d'applications (application programming interface) ; il s'agit d'une interface permettant au site Web (ou à l'application mobile) de communiquer avec la logique backend. En termes simples, c'est comme un messager qui prend une demande des utilisateurs et envoie la demande au système backend. Une fois que le système backend aura répondu, il transmettra ensuite cette réponse aux utilisateurs. Une métaphore pour cela est un serveur/serveuse, qui peut comprendre les commandes de différents clients. Ils serviront ensuite d'intermédiaire entre les clients et les chefs en cuisine

## RESTful 

APIREST signifie Representational State Transfer. Il a été défini pour la première fois dans la thèse du Dr Roy Fielding (Styles architecturaux et conception d'architectures logicielles basées sur le réseau) en 2000. Cette thèse est considérée comme la bible du domaine Web. REST n'est pas une norme ou un protocole ; cela ressemble plus à un style architectural logiciel. De nombreux ingénieurs suivent ce style architectural pour créer leurs applications, comme eBay, Facebook et Google Maps. Ces applications Web servent d'énormes quantités de trafic chaque seconde, vous pouvez donc voir que REST est vraiment un style d'architecture évolutif. Et quand on dit API RESTful, on fait référence à une API conforme aux contraintes/principes REST

## Contraintes/principes REST 

Il existe cinq contraintes/principes importants pour le style d'architecture REST : 

*  Client-serveur : il existe une interface entre le client et le serveur. Le client et le serveur communiquent via cette interface et sont indépendants l'un de l'autre. Chaque côté peut être remplacé tant que l'interface reste la même. Les demandes proviennent toujours du côté client.

* Sans état (Stateless): il n'y a pas de concept d'état pour une demande. Chaque demande est considérée comme indépendante et complète. Il n'y a aucune dépendance vis-à-vis de la demande précédente ni de dépendance vis-à-vis d'une session pour maintenir l'état de la connexion. 

* Cacheable : les éléments peuvent être mis en cache sur le serveur ou côté client pour améliorer les performances.

* Système en couches : il peut y avoir plusieurs couches dans le système, et le but ici est de masquer la logique/les ressources réelles. Ces couches peuvent exécuter différentes fonctions, telles que la mise en cache et le cryptage

* Interface uniforme : L'interface reste la même. Cela permet de découpler la logique client et serveur

## Protocole HTTP

Pour mieux comprendre ce qu'est REST et s'assurer que nous implémentons le style REST, nous pouvons simplement parler du protocole HTTP. HTTP est une implémentation du style d'architecture REST. C'est l'abréviation de HyperText Transfer Protocol et c'est le protocole standard utilisé sur le Web mondial. Nous l'utilisons tous les jours pour parcourir différents sites Web. C'est pourquoi tous les sites Web que nous visitons sont préxés par http. Dans le protocole HTTP, il existe différents types de méthodes de demande de service. Chaque méthode de demande de service a une dénition spéciale qui lui est propre. 

Lorsque l'interface frontend interagit avec l'API backend via une URL, ils doivent, en même temps, définir la méthode HTTP pour cette requête. Différentes méthodes HTTP sont comme différents compteurs de services. Par exemple, la lecture et la création de données sont des services complètement différents, elles doivent donc être gérées par des compteurs de services différents, ce qui signifie des méthodes HTTP différentes.

* GET : pour lire des données
* POST : pour créer des données
* PUT : pour mettre à jour des données en remplaçant complètement les données par nouveau contenu
* PATCH : pour mettre à jour des données, mais en modifiant partiellement quelques attributs
* DELETE : pour supprimer des données 

En termes simples, différentes méthodes HTTP sont comme les verbes pour l'API REST. Ils sont utilisés pour effectuer différentes actions sur le même ensemble de données.

## Codes d'état HTTP

Un code d'état HTTP est un code renvoyé dans le protocole HTTP. Il est généralement caché aux utilisateurs, vous n'avez donc probablement pas réalisé qu'il existait. En fait, chaque réponse HTTP du serveur contient un code d'état. Et lorsque nous construisons notre API RESTful, nous devons nous conformer au protocole HTTP. Le code d'état aide le client frontal à comprendre l'état de sa demande, c'est-à-dire s'il s'agit d'un succès ou d'un échec. Par exemple, il peut y avoir une demande du client concernant la création d'un enregistrement dans la base de données principale. Dans ce cas, une fois que l'enregistrement de la base de données a été créé avec succès, le serveur doit renvoyer un code d'état HTTP 201 (Créé). S'il y a une erreur (telle qu'une erreur de syntaxe dans le document JSON), le serveur doit plutôt renvoyer un code d'état HTTP 400 (Bad Request).

## Codes d'état HTTP couramment utilisés 

Examinons quelques codes d'état couramment utilisés. Ils sont les suivants :
*  200 OK signifie que la demande a abouti. La demande peut être un GET, PUT ou PATCH.

* 201 Créé signifie que la demande POST a réussi et qu'un enregistrement a été créé.

* 204 Aucun contenu signifie que la demande DELETE a réussi.

* 400 Mauvaise demande signifie qu'il y a quelque chose qui ne va pas avec la demande du client. Par exemple, il y a une erreur de syntaxe dans le format JSON.

* 401 Unauthorized signifie qu'il manque des détails d'authentification à la demande du client.

* 403 Forbidden signifie que la ressource demandée est interdite.

* 404 Not Found signifie que la ressource demandée n'existe pas

```python
from flask import Flask, jsonify, request
from http import HTTPStatus

app = Flask(__name__)

recipes = [
    {
        'id': 1,
        'name': 'Egg Salad',
        'description': 'This is a lovely egg salad recipe.'
    },
    {
        'id': 2, 'name': 'Tomato Pasta',
        'description': 'This is a lovely tomato pasta recipe.'
    }
]


@app.route('/recipes', methods=['GET'])
def get_recipes():
    return jsonify({'data': recipes})


@app.route('/recipes/<int:recipe_id>', methods=['GET'])
def get_recipe(recipe_id):
    recipe = next((recipe for recipe in recipes if recipe['id'] == recipe_id), None)

    if recipe:
        return jsonify(recipe)

    return jsonify({'message': 'recipe not found'}), HTTPStatus.NOT_FOUND


@app.route('/recipes', methods=['POST'])
def create_recipe():
    data = request.get_json()

    name = data.get('name')
    description = data.get('description')

    recipe = {
        'id': len(recipes) + 1,
        'name': name,
        'description': description
    }

    recipes.append(recipe)

    return jsonify(recipe), HTTPStatus.CREATED


@app.route('/recipes/<int:recipe_id>', methods=['PUT'])
def update_recipe(recipe_id):
    recipe = next((recipe for recipe in recipes if recipe['id'] == recipe_id), None)

    if not recipe:
        return jsonify({'message': 'recipe not found'}), HTTPStatus.NOT_FOUND

    data = request.get_json()

    recipe.update(
        {
            'name': data.get('name'),
            'description': data.get('description')
        }
    )

    return jsonify(recipe)


if __name__ == '__main__':
    app.run()
```
